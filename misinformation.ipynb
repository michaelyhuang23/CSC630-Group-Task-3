```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import altair as alt
import numpy as np
import pandas as pd
import datetime


def cond(word):
    if not (not word.isalnum() or 'donald' in word or word in st_words):
        return True
    else:
        return False

data = pd.read_csv("Fake Email.csv")
cc = Counter()
ndates = []
cdates = []
for i, row in data.iterrows():
    txt = row['text'].lower().split()
    date = row['date']
    try:
        ndates.append(date.split(', ')[1]+" "+date.split(' ')[0])
    except:
        ndates.append('2015 Jan')
        continue
        
    txt = [xt for xt in txt if cond(xt)]
    for word in txt:
        cc[word]+=1
        
cdates = ndates.copy()

for i in range(len(ndates)):
    ndates[i] = ndates[i].replace('December', 'Dec')
    ndates[i] = ndates[i].replace('August', 'Aug')
    ndates[i] = ndates[i].replace('February', 'Feb')
    ndates[i] = ndates[i].replace('January', 'Jan')
    ndates[i] = ndates[i].replace('October', 'Oct')
    ndates[i] = ndates[i].replace('June', 'Jun')
    ndates[i] = ndates[i].replace('July', 'Jul')
    ndates[i] = ndates[i].replace('September', 'Sep')
    ndates[i] = ndates[i].replace('March', 'Mar')
    ndates[i] = ndates[i].replace('April', 'Apr')
    ndates[i] = ndates[i].replace('November', 'Nov')
    
    cdates[i] = ndates[i].replace('Dec', '12')
    cdates[i] = cdates[i].replace('Nov', '11')
    cdates[i] = cdates[i].replace('Oct', '10')
    cdates[i] = cdates[i].replace('Sep', '9')
    cdates[i] = cdates[i].replace('Aug', '8')
    cdates[i] = cdates[i].replace('Jul', '7')
    cdates[i] = cdates[i].replace('Jun', '6')
    cdates[i] = cdates[i].replace('May', '5')
    cdates[i] = cdates[i].replace('Apr', '4')
    cdates[i] = cdates[i].replace('Mar', '3')
    cdates[i] = cdates[i].replace('Feb', '2')
    cdates[i] = cdates[i].replace('Jan', '1')
    #print(cdates[i], ndates[i])
    cdates[i] = datetime.datetime.strptime(cdates[i], "%Y %m").date()


data['date']=cdates


front = cc.most_common()[:20]
print(cc.most_common()[:20])
front = [fr[0] for fr in front]

data=data.sort_values('date')
cdates = data['date']
fdates = []
for date in cdates:
    if len(fdates)==0 or fdates[-1]!=date:
        fdates.append(date)

print(data.head())

counter = {}
for i, row in data.iterrows():
    c = Counter()
    txt= row['text'].lower().split()
    date = str(row['date'])
    txt = [xt for xt in txt if cond(xt)]
    for word in txt:
        if word in front:
            c[word]+=1
    if not date in counter: counter[date] = Counter()
    counter[date].update(c)

src = []
for date in fdates:
    c = counter[str(date)]
    for k, v in c.items():
        src.append((str(date)[:-3], k, v))

df = pd.DataFrame(src, columns=['date','word','count'])

# colors = ['red', 'orange', 'blue']
# cols = []
# for i, row in df.iterrows():
# 	cols.append(colors[front.index(row['word'])])

# df['col'] = cols
df = df[['date','word','count']]


print(df)
```

```python
import pandas as pd
emails = pd.read_csv("Fake Email.csv")
cctit = Counter()
cctxt = Counter()
for i, row in emails.iterrows():
    tit = row['title'].lower()
    txt = row['text'].lower()
    tit = tit.split()
    txt = txt.split()
    tit = [it for it in tit if not it in st_words]
    tit = [it for it in tit if not (not it.isalnum() or 'donald' in it)]
    txt = [xt for xt in txt if not xt in st_words]
    txt = [xt for xt in txt if not (not xt.isalnum() or 'donald' in xt)]
    ctit = Counter(tit)
    ctxt = Counter(txt)
    cctit.update(ctit)
    cctxt.update(ctxt)
```
