```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import altair as alt
import numpy as np
import pandas as pd

data = pd.read_csv("Fake Email.csv")
cc = Counter()
ndates = []
for i, row in data.iterrows():
    txt = row['title']
    date = row['date']
    try:
        ndates.append(date.split(' ')[0]+" "+date.split(', ')[1])
    except:
        ndates.append('Jan 2015')
        continue
    for word in txt.split():
        cc[word.lower()]+=1
        
for i in range(len(ndates)):
    ndates[i] = ndates[i].replace('December', 'Dec')
    ndates[i] = ndates[i].replace('August', 'Aug')
    ndates[i] = ndates[i].replace('February', 'Feb')
    ndates[i] = ndates[i].replace('January', 'Jan')
    ndates[i] = ndates[i].replace('October', 'Oct')
    ndates[i] = ndates[i].replace('June', 'Jun')
    ndates[i] = ndates[i].replace('September', 'Sep')
    ndates[i] = ndates[i].replace('March', 'Mar')
    ndates[i] = ndates[i].replace('November', 'Nov')

data['date']=ndates

front = cc.most_common()[:3]
print(front)
front = [fr[0] for fr in front]

counter = {}
for i, row in data.iterrows():
	c = Counter()
	txt= row['title'].lower()
	date = row['date']
	for word in txt.split():
		if word in front:
			c[word]+=1
	if not date in counter: counter[date] = Counter()
	counter[date].update(c)

src = []
for date in counter.keys():
	c = counter[date]
	for k, v in c.items():
		src.append((date, k, v, ''))

df = pd.DataFrame(src, columns=['date','word','count','col'])

print(df)

# colors = ['red', 'orange', 'blue']
# cols = []
# for i, row in df.iterrows():
# 	cols.append(colors[front.index(row['word'])])

# df['col'] = cols
print(df)
df = df[['date','word','count']]
    
```

```python
alt.Chart(df).mark_line().encode(
    x='date',
    y='count',
    color=alt.Color('word', legend=None)
).properties(title={
      "text": ["The Three Most Common Words People Tweeted"]})
```
